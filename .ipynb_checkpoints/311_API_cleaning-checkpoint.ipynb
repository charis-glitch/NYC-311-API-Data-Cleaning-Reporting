```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pandas import json_normalize

%matplotlib inline
sns.set(style="whitegrid")

output_folder = "outputs"
os.makedirs(output_folder, exist_ok=True)
```

```python
# Fetch recent 5000 records from NYC 311 Service Requests (SODA API) and convert the data into JSON format
api_url = "https://data.cityofnewyork.us/resource/erm2-nwe9.json"
params = {"$limit": 5000, "$order": "created_date DESC"}
resp = requests.get(api_url, params=params, timeout=30)
resp.raise_for_status()
data = resp.json()
```

```python
# Convert nested JSON to DataFrame
df = pd.json_normalize(data)

print("Raw data preview:")
display(df.head())
print("\nData info:")
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
print("\nDuplicates:")
print(df.duplicated().sum())
```

```python
# Remove duplicated rows & columns
df.drop_duplicates(inplace=True)
df = df.T.drop_duplicates().T

# Standardize column names
df.columns = df.columns.str.replace(r'[._]', ' ', regex=True).str.title()

# Convert date column   
for col in ['Created Date', 'Closed Date', 'Resolution Action Updated Date', 'Due Date']:
    df[col] = pd.to_datetime(df[col],  format="%Y-%m-%d", errors='coerce', utc=True).dt.tz_localize(None)

# Drop rows with no Unique Key and no Created Date (likely invalid)
mask_invalid = df['Unique Key'].isna() & df['Created Date'].isna()
removed = mask_invalid.sum()
if removed:
    df = df.loc[~mask_invalid]
    print(f"Removed {removed} rows with no Unique Key and no Created Date")

# Handle missing values
for col in df.columns[df.isnull().sum() > 0]:
    df[col] = df[col].fillna('Unknown')

# Check result
print("\n--- Data Cleaning Complete ---")
df.info()
print("Cleaned data preview:")
display(df.head())
print("Data shape after cleaning:", df.shape)
```

```python
# Summaries
# Top 20 complaint types
df['Complaint Type'] = df['Complaint Type'].str.title()
top_complaints = df['Complaint Type'].value_counts().reset_index().head(20)

# Borough counts
df['Borough'] = df['Borough'].str.title()
borough_counts = df['Borough'].value_counts().reset_index()
```

```python
# Top 20 complaint types (Bar Chart) 
plt.figure(figsize=(10, 6))
sns.barplot(x='count', y='Complaint Type', data=top_complaints, hue='Complaint Type', palette='viridis', legend=False)
plt.title('Top 20 complaint types')
plt.xlabel('count')
plt.ylabel('Complaint Type')
plt.tight_layout()
fn = os.path.join(output_folder, 'top_complaints.png')
plt.savefig(fn)
plt.show()

# Borough counts (Bar Chart)
plt.figure(figsize=(10, 6))
sns.barplot(x='Borough', y='count', data=borough_counts, hue='count', palette='magma', legend=False)
plt.title('Borough counts')
plt.xlabel('Borough')
plt.ylabel('count')
plt.tight_layout()
fn = os.path.join(output_folder, 'borough_counts.png')
plt.savefig(fn)
plt.show()
```

```python
# Save cleaned data and summaries back to the Excel file
cleaned_xlsx = os.path.join(output_folder, 'cleaned_311_data.xlsx')

with pd.ExcelWriter(cleaned_xlsx, engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='Clean Data', index=False)
    top_complaints.to_excel(writer, sheet_name='Analysis & Summary', startrow=0, startcol=0, index=False)
    borough_counts.to_excel(writer, sheet_name='Analysis & Summary', startrow=25, startcol=0, index=False)

print("\n*** Successfully saved cleaned data and summaries to an Excel file. ***")
```
